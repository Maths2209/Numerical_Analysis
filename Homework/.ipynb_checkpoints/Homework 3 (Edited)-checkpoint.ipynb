## Q1(a)

Let $f'(x_j) = af(x_j + 2h) + bf(x_j + h) + cf(x_j) + df(x_j - h)$.
Taylor expansion of $f(x_j \pm h)$ at $x$:
$$f(x_j \pm h) = f(x_j) \pm f'(x_j)h + \frac{f''(x_j)h^2}{2!} \pm \frac{f''(x_j)h^3}{3!} + \mathcal{O}(h^4).$$
Taylor expansion of $f(x_j + 2h)$ at $x$:
$$f(x_j + 2h) = f(x_j) + 2f'(x_j)h + \frac{4f''(x_j)h^2}{2!} \pm \frac{8f''(x_j)h^3}{3!} + 16\mathcal{O}(h^4).$$
Combining them, we have,
$$f'(x_j) = (a+b+c+d)f(x_j) + (2a+b-d)(f'(x_j)h + (4a+b+d)\frac{f''(x_j)h^2}{2!} + (8a+b-d)\frac{f''(x_j)h^3}{3!} + (16a+b+d)\mathcal{O}(h^4)$$

By comparing coefficients, we have the system of linear equations:
\begin{align*}
&a+b+c+d = 0 \\
&2a+b-d = \frac{1}{h} \\
&4a+b+d = 0 \\
&8a+b-d = 0 \\
\end{align*}
Solving in terms of $h$, we get 
$$a = -\frac{1}{6h}, b = \frac{1}{h}, c = -\frac{1}{2h}, -\frac{1}{3h}$$
Therefore, $$f'(x_j) = -\frac{1}{6h}f(x_j + 2h) + \frac{1}{h}f(x_j + h) -\frac{1}{2h}f(x_j) -\frac{1}{3h}f(x_j - h)$$

This method is Order 4; error $= \mathcal{O}(h^4)$.

## Q1(b)

### Define Finite Differences methods:

```python
def forward_d(f,x,h): #Forward-Difference, Order 1
    return (f(x+h)-f(x))/h

def centered_d(f,x,h): #Centered-Difference, Order 2
    return (f(x+h)-f(x-h))/(2*h)

def fourpoint_d(f,x,h): #4-Point, Finite-Difference, Order 4
    return (-(1/(6*h))*f(x+2*h)+(1/h)*f(x+h)-(1/(2*h))*f(x)-(1/(3*h))*f(x-h))
```

### Define Lagrange Interpolation Differentiation:

```python
def L(xvals,kk,x):
    value = np.ones(x.size)
    n = xvals.size
    for ii in range(0,n):
        if ii != kk:
            value *= (x-xvals[ii])/(xvals[kk]-xvals[ii])
    return value

def LpSum(xvals,kk,x):
    lval = np.zeros(x.size)
    n = xvals.size
    for ii in range(0,n):
        if ii != kk:
            lval += 1/(x-xvals[ii])
    return lval

def p_LagrangeP(xvals,fvals,x):
    n = fvals.size
    pLagrange = np.zeros(x.size)
    for kk in range(0,n):
        pLagrange += fvals[kk]*L(xvals,kk,x)*LpSum(xvals,kk,x)
    return pLagrange

def Lagrange_d(f,x,h): #Lagrange Interpolation Derivative.
    xvals=np.array([x-h,x,x+h])
    yvals=f(xvals)
    xa = np.array([x+(10**(-13))]) #Offset the point slightly to prevent a division by zero when evaluating at the interpolation point
    return p_LagrangeP(xvals,yvals,xa)[0]
```

### Define Cubic Spline Interpolation Differentiation:

```python
def cubic_spline_coeffs(xvals,yvals):

    n = xvals.size - 1
    h = xvals[1:]-xvals[0:-1]

    d0 = np.hstack([1,2*(h[0:-1]+h[1:]),1])
    d1 = np.hstack([0,h[1:]])
    dm1 = np.hstack([h[0:-1],0])
    # create the "A" matrix by using the np.diag command
    A = np.diag(d0) + np.diag(d1,1)+ np.diag(dm1,-1)

    
    # create the right-hand-side of A*x = b
    # recall that a_j = y_j
    aVec = yvals
    rhs = np.hstack([0,(3./h[1:])*(aVec[2:]-aVec[1:-1])-(3/h[0:-1])*(aVec[1:-1]-aVec[0:-2]),0])
    
    # use the linalg.solve command to solve Ax = b  
    cVec = np.linalg.solve(A,rhs)

    
    # use the remaining formula to determine d_j and b_j
    dVec = (cVec[1:]-cVec[0:-1])/(3*h)
    bVec = 1/h*(aVec[1:] - aVec[0:-1]) - h/3*(2*cVec[0:-1]+cVec[1:])
    
    # stack all of the coefficients into a matrix so that the coefficients are in the form 
    #            a_0, a_1, ...
    #            b_0, b_1, ...
    #            c_0, c_1, ...
    #            d_0, d_1, ...
    
    SCoeffs = np.vstack([aVec[0:n], bVec[0:n], cVec[0:n], dVec[0:n]])
    return SCoeffs

def cubic_spline_fp(x, coeffs, xvals):
    n = len(coeffs[0])
    ans = 0
    for ii in range(1,4):
        for jj in range(0,n):
            ans += coeffs[ii][jj]*ii*((x-xvals[jj])**(ii-1))*((xvals[jj]<x) and (x<=xvals[jj+1]))
    return ans

def cubic_spline_d(f,x,h): #Cubic Spline Interpolation Derivative.
    xvals=np.array([x-h,x,x+h])
    yvals=f(xvals)
    coeffs=cubic_spline_coeffs(xvals,yvals)
    return cubic_spline_fp(x,coeffs,xvals)
```

```python
f = lambda x: (np.sin(x) + 1)**(np.sin(np.cos(x)))

print("Forward-Difference, Order 1; x = 1, h = 0.1")
print(forward_d(f,1,0.1))
print()
print("Centered-Difference, Order 2; x = 1, h = 0.1")
print(centered_d(f,1,0.1))
print()
print("4-Point, Finite-Difference, Order 4; x = 1, h = 0.1")
print(fourpoint_d(f,1,0.1))
print()
print("Lagrange Interpolation Derivative; x = 1, h = 0.1")
print(Lagrange_d(f,1,0.1))
print()
print("Cubic Spline Interpolation Derivative; x = 1, h = 0.1")
print(cubic_spline_d(f,1,0.1))
print()
print("Cubic Spline Interpolation Derivative; x = 1, h = 0.1")
print(cubic_spline_d(f,1,0.1))
print()
print("Actual value (from WolframAlpha)")
print(-0.3965403874194624) #It appears that the 4-Point, Finite-Difference method is the most accurate.
                           #The Forward-Difference is the least accurate (as would be expected).
```

For $h \leq 10^{-4}$, rounding error starts to dominate, though this phenonemon is also dependent on the method used: for $10^{-8} \leq h \leq 1$, rounding error does not seem to significantly affect the Forward Difference method.

## Q3(b)

Suppose that $u''(x) = au(x-h) + bu(x) +cu(x+h)$. We will Taylor expand $u(x-h)$ and $u(x+h)$ at point $x$.

\begin{align*}
au(x-h) &= a\left(u(x) + u'(x)(x-h-x) + \frac{u''(x)}{2!}(x-h-x)^2 + \frac{u^{(3)}(x)}{3!}(x-h-x)^3 + \frac{u^{(4)}(\xi_1)}{3!}(x-h-x)^4\right) \\
&= a\left(u(x) - u'(x)h + \frac{u''(x)}{2!}h^2 - \frac{u^{(3)}(x)}{3!}h^3 + \frac{u^{(4)}(\xi_1)}{3!}h^4\right)
\end{align*}
\begin{align*}
cu(x+h) &= c\left(u(x) + u'(x)(x+h-x) + \frac{u''(x)}{2!}(x+h-x)^2 + \frac{u^{(3)}(x)}{3!}(x+h-x)^3 + \frac{u^{(4)}(\xi_2)}{3!}(x+h-x)^4\right) \\
&= c\left(u(x) + u'(x)h + \frac{u''(x)}{2!}h^2 + \frac{u^{(3)}(x)}{3!}h^3 + \frac{u^{(4)}(\xi_2)}{3!}h^4\right)
\end{align*}
Therefore,
$$u''(x) = (a+b+c)u(x) + (c-a)u'(x)h + (a+c)\frac{u''(x)}{2!}h^2 + (c-a)\frac{u^{(3)}(x)}{3!}h^3 + (c+a)\mathcal{O}(h^4).$$
By comparing coefficients, $\displaystyle a+b+c=0, c-a=0, (a+c)\frac{h^2}{2!}=1$. Therefore, 
$$c = a = \frac{1}{h^2}, \hspace{1pc} b = -2a = -\frac{2}{h^2}.$$
Since $\displaystyle(c+a)\mathcal{O}(h^4) = \frac{2}{h^2}\mathcal{O}(h^4) = \mathcal{O}(h^2)$, the order of the method is $\mathcal{O}(h^2)$.

## Q3(c)

We will cast the problem into the from $A\vec{U} = \vec{G}$ using the finite difference formula,
$$u''(x) = \frac{1}{h^2}u(x-h) - \frac{2}{h^2}u(x) + \frac{1}{h^2}u(x+h).$$

$$\therefore A = \frac{1}{h^2}\left(
\begin{array}{ccccccccc}
-2 & 1 & 0 & 0 & 0 & \ldots & 0 & 0 & 0\\
 1 & -2 & 1 & 0 & 0 & \ldots & 0 & 0 & 0\\
 0 & 1 & -2 & 1 & 0 & \ldots & 0 & 0 & 0\\
 \vdots & \vdots & \vdots & \vdots & \vdots & \ddots & \vdots & \vdots & \vdots \\
 0 & 0 & 0 & 0 & 0 &\ldots & 1 & -2 & 1 \\
 0 & 0 & 0 & 0 & 0 & \ldots & 0 & 1 & -2
 \end{array} \right) \hspace{2pc}
 \vec{U} = \left(
 \begin{array}{c}
 u(x_1) \\
 u(x_2) \\
 \vdots \\
 u(x_n)
 \end{array} \right) \hspace{2pc}
 \vec{G} = \left(
 \begin{array}{c}
 u''(x_1) - \displaystyle \frac{1}{h^2}u(x_0) \\
 u''(x_2) \\
 \vdots \\
 u''(x_{n-1}) \\
 u''(x_n) - \displaystyle \frac{1}{h^2}u(x_{n+1})
 \end{array} \right)$$

## Q3(d)

```python
def solve_for_U(xvals,upp):
    h = xvals[2:]-xvals[1:-1] #h_j := x_{j+1} - x_j.
                              #h_j's definition results in x_0 being ignored, whereas x_n is used to calculated h_{n-1}.
                              #The answer did specify a formula for h, but this definition is more flexible (allows for non-equispaced points).
    d0 = -2*(1/h**2)
    d1 = 1/h[1:]**2
    dm1 = 1/h[0:-1]**2
    # create the "A" matrix by using the np.diag command
    A = sp.diags(d0) + sp.diags(d1,1)+ sp.diags(dm1,-1)
    
    
    # create the "b" matrix by applying the function to x_i.
    uppvals = np.hstack([upp(xvals[1:-2]),upp(xvals[-2])-1/(h[-2]**2)])
    # Ax = b, solve for x.
    uvals = spsolve(A,uppvals) #I'm using scipy sparse matrice solver else I'll run into memory problems later.
    
    return uvals
```

```python
u_0, u_n = 0, 1 #Boundary conditions
upp = lambda x: -(np.pi**2)*np.sin(np.pi*x)
xvals = np.linspace(0,1,10) #xvals includes x_0 for graphing purposes, but x_0 is not used in the function.

incomplete_yvals = solve_for_U(xvals,upp) #This is an array of u_1,...,u_{n-1}.
#Add u_0 and u_n into yvals.
yvals = np.insert((np.insert(incomplete_yvals,len(incomplete_yvals),u_n)),0,u_0)

u = lambda x: x + np.sin(x*np.pi)
yvals_actual = u(xvals)

plt.grid(True)
plt.plot(xvals,yvals) #The numerical solution.
plt.plot(xvals,yvals_actual) #The actual solution.
# They are very similar, even though h is a mere 0.1!!
plt.legend(["Numerical solution","Actual solution"])
plt.grid()
plt.title('Plot of the actual and numerical solutions to the ODE',fontsize=18)
plt.xlabel('$x$',fontsize=14)
plt.ylabel('$y$',fontsize=14)
plt.grid(True)
```

## 3(e)

```python
def error(h, upp, u):
    xvals = np.linspace(0,1,int(1/h))
    
    u_0, u_n = 0, 1
    
    incomplete_yvals = solve_for_U(xvals,upp)
    yvals = np.insert((np.insert(incomplete_yvals,len(incomplete_yvals),u_n)),0,u_0)
    
    yvals_actual = u(xvals)
    diff = np.abs(yvals_actual - yvals)
    err = np.max(diff) #The error is the max difference between yvals and yvals_actual.
    return err
```

```python
upp = lambda x: -(np.pi**2)*np.sin(np.pi*x)
u = lambda x: x + np.sin(x*np.pi)

n = 5
h = np.zeros(n)
for ii in range(0,n):
    h[ii] = 0.1**(1+ii)

results = np.zeros(n)
for ii in range(0,n):
    results[ii] = error(h[ii],upp,u)

plt.loglog(h,results)
plt.title('Log-Log Plot of Error for various $h$',fontsize=18)
plt.xlabel('$\log(h)$',fontsize=14)
plt.ylabel('$\log$(Error)',fontsize=14)
plt.grid(True)
```

$\ln(error)$ is directly proportional to $\ln(h)$.
